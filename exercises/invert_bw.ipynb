```python
from skimage import io
from skimage.color import rgb2gray
import matplotlib.pyplot as plt

img_dir = "data/"
img_name = "hi.png"
input_img = io.imread(img_dir + img_name)
input_img = rgb2gray(input_img)
io.imshow(input_img)
io.show()
```

```python
inverted = 1-input_img
io.imshow(inverted)
io.show()
```

```python
fig, (ax0, ax1) = plt.subplots(ncols=2) # figsize = (10,10)
ax0.set_title("original")
ax0.imshow(input_img, cmap='gray')
ax0.axis('off')

ax1.set_title("dingal's request")
ax1.imshow(inverted, cmap='gray')
ax1.axis('off')

fig.tight_layout()
io.show()
```
