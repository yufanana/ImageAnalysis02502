```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio

in_dir = 'data/'
in_file = 'ex6_ImagData2Load.mat'
data = sio.loadmat(in_dir + in_file)
ImgT1 = data['ImgT1']
ImgT2 = data['ImgT2']
ROI_GM = data['ROI_GM']
ROI_WM = data['ROI_WM']
```

```python
fig, axs = plt.subplots(2,2)
axs[0,0].imshow(ImgT1, cmap='gray')
axs[0,0].set_title('ImgT1')
axs[0,0].axis('off')
axs[0,1].imshow(ImgT2, cmap='gray')
axs[0,1].set_title('ImgT2')
axs[0,1].axis('off')
axs[1,0].hist(ImgT1)
axs[1,0].set_title('Histogram of ImgT1')
axs[1,1].hist(ImgT2)
axs[1,1].set_title('Histogram of ImgT2')

# axs[4].imshow(ImgT1, cmap='gray')
# axs[5].imshow(ImgT2, cmap='gray')
plt.show()

## OR 
# plt.subplot(1,2,1)
# plt.imshow(ImgT1, cmap='gray')
# plt.title('ImgT1')
# plt.subplot(1,2,2)
# plt.imshow(ImgT2, cmap='gray')
# plt.title('ImgT2')
# plt.show()
```
